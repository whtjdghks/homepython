```python
stack_data=df1.drop(columns=['순번','상품코드']).set_index('상품명').stack()
```

```python
df2=pd.DataFrame(stack_data).reset_index()
```

```python
cond1 = (df2['상품명'].isnull())
cond2=(df2['상품명']!='합 계')
cond3 = (df2['level_1']!=' 합계') #공백 주의
df3=df2.loc[(~cond1)&(cond2)&(cond3)]
df4=df3.rename(columns={'level_1':'점포명', 0:'판매량'})
#이게 전부 상품명이 미싱된 값들의 반대의 의미를 가진 ~
# 미싱된 값들이 전부 날아간다.

#결과는 상품명 점포명 판매량만 남겨서 추출한결과임
```

```python
df4['상품명'].unique()
```

```python
# 타입별 분류
# 분류시키기~
def func1(row):
    if 'CAT' in row:
        return 'C Type'
    elif 'STL' in row:
        return 'S Type'
    elif 'BW' in row:
        return 'B Type'
    else:
        return row

```

```python
df4['제품군']=df4['상품명'].apply(func1)
df4
```
