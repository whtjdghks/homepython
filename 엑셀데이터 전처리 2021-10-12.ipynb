# 이제 df5 와 df6를 붙여줘야 함!

# 붙이는 방법 4가지
* Table Join <br>
pd.merge(데이터,데이터,how='방법')
- right : 오른쪽값이 기준이 된다.
- left : 왼쪽 테이블 기준으로 붙인다. 즉 왼쪽에 있는 데이터에 있으면 ok 없으면 NaN값으로 처리된다.
- inner : 키값에 중복된 데이터만 가져와서 붙인다.
- outer : 중복되지 않은 데이터도 다 가져와서 붙인다. 하지만 있는 데이터만 가져와서 붙인다. NaN 발생

```python
# right 
df7=pd.merge(df6,df5,how='right',on='제품명')# on은 기준이 되는 데이터를 의미한다.

df7.to_excel('total_data.xlsx') # 전처리된 위의 데이터를 엑셀데이터화
```

```python
df7['날짜(전처리)']=pd.to_datetime(df7['날짜'])
```

```python
df7['공급월']=df7['날짜(전처리)'].dt.month
```

```python
pd.pivot_table(data=df7,index='공급월',
               values='재고량',aggfunc='sum').to_excel('result1.xlsx')
```
