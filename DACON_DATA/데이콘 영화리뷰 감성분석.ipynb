```python
train=train['document'].str.replace('[^ㄱ-ㅎ ㅏ-ㅣ가-힣]','',regex=True)
```

```python
train.count()
```

```python
okt=Okt()
```

```python
# 토큰화
stop = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']
```

```python
train_s = []

for ts in tqdm(train['document']):
   
    token_ts = okt.morphs(ts,stem=True)
    stop_ts = [spts for spts in token_ts if not spts in stop]
    train_s.append(stop_ts)
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_s)
```
