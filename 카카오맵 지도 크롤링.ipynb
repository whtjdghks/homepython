```python

import re
from selenium.webdriver.common.action_chains import ActionChains

from bs4 import BeautifulSoup     
from selenium import webdriver
import time
import pandas as pd
# path = "C:/chrome/chromedriver.exe"
# driver = webdriver.Chrome(path)
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')        


df = pd.DataFrame(columns=['업종','가게이름','평가점수','평가인원', 
                           '리뷰갯수', '주소', "전화", "영업시간","링크주소"])
lili= soup.find_all("li","PlaceItem clickArea")

############ AS 2-1단
for a in range(len(lili)):

    li= soup.find_all("li","PlaceItem clickArea")[a]
    upjong = li.find("span","subcategory clickable").text #업종
    shop = li.find("a","link_name").text               #가게이름
    point = li.find("em","num").text                   #평가점수
    person = li.find("a","numberofscore").text[:-1]  #몇명평가
    reviews2 = li.find("a","review").text[3:]         #리뷰-블로그
    add1 = li.find("div","addr")
    add = add1.find_all("p")[0].text                     #주소
    phone1 = li.find("div","contact clickArea")
    phone = phone1.find("span").text                     #전화번호
    selltime = li.find("div","openhour").text.replace("\n","")[3:] #영업시간    
    b_link = li.find_all("a")[5].get('href')    #블로그 링크
    ############### AS 2-2단

    df=df.append({'업종':upjong,'가게이름':shop,'평가점수':point,'평가인원':person, 
                  '리뷰갯수':reviews2, '주소':add, "전화":phone, "영업시간":selltime,"링크주소":b_link}, ignore_index=True)

df.to_csv(f"~/Desktop/다음지도-검색.csv", index=False, mode='w', encoding='utf-8-sig') 
print("1 번째 파싱")
try:
    time.sleep(1)
    ###################### AS 2-3단
    element1 = driver.find_element_by_id("info.search.place.more")
    ActionChains(driver).move_to_element(element1).click(element1).perform()
    # 더보기 누르기 1초 넣기


    try:
        for b in range(3,6,1):
            time.sleep(1)
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')

            lili= soup.find_all("li","PlaceItem clickArea")[1]
            ############# AS 3-1단
            for a in range(len(lili)):
                selltime="안적어놨어"
                li= soup.find_all("li","PlaceItem clickArea")[a]
                upjong = li.find("span","subcategory clickable").text #업종
                shop = li.find("a","link_name").text               #가게이름
                point = li.find("em","num").text                   #평가점수
                person = li.find("a","numberofscore").text[:-1]      #몇명평가
                reviews2 = li.find("a","review").text[3:]           #리뷰-블로그
                add1 = li.find("div","addr")
                add = add1.find_all("p")[0].text                     #주소
                phone1 = li.find("div","contact clickArea")
                phone = phone1.find("span").text                     #전화번호
                selltime = li.find("div","openhour").text.replace("\n","")[3:] #영업시간
                b_link = li.find_all("a")[5].get('href')                #블로그 링크
                ########### AS 3-2단
                df=df.append({'업종':upjong,'가게이름':shop,'평가점수':point,'평가인원':person, 
                              '리뷰갯수':reviews2, '주소':add, "전화":phone, "영업시간":selltime,"링크주소":b_link}, ignore_index=True)


            df.to_csv(f"~/Desktop/다음지도-검색.csv", index=False, mode='w', encoding='utf-8-sig') 
            print( b-1, "번째 파싱")        

            time.sleep(1)
            ############## AS 3-3단
            
            
            
            element1 = driver.find_element_by_id("info.search.page.no{}".format(b))
            ActionChains(driver).move_to_element(element1).click(element1).perform()
    except:
        pass
    print("5 번째 까지 했슈")    

    time.sleep(1)
    ############### AS 4-1 단       info.search.page.next
    element1 = driver.find_element_by_id("info.search.page.next")
    ActionChains(driver).move_to_element(element1).click(element1).perform()


    for c in range(2,7,1):
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        lili= soup.find_all("li","PlaceItem clickArea")[1]
        ################## AS 4-2단
        for a in range(len(lili)):
            selltime="안적어놨어"
            li= soup.find_all("li","PlaceItem clickArea")[a]
            upjong = li.find("span","subcategory clickable").text #업종
            shop = li.find("a","link_name").text               #가게이름
            point = li.find("em","num").text                   #평가점수
            person = li.find("a","numberofscore").text[:-1]    #몇명평가
            reviews2 = li.find("a","review").text[3:]          #리뷰-블로그
            add1 = li.find("div","addr")
            add = add1.find_all("p")[0].text                     #주소
            phone1 = li.find("div","contact clickArea")
            phone = phone1.find("span").text                     #전화번호
            selltime = li.find("div","openhour").text.replace("\n","")[3:] #영업시간
            ################## AS 4-3단

            df=df.append({'업종':upjong,'가게이름':shop,'평가점수':point,'평가인원':person, 
                          '리뷰갯수':reviews2, '주소':add, "전화":phone, "영업시간":selltime,"링크주소":b_link}, ignore_index=True)


        df.to_csv(f"~/Desktop/다음지도-검색.csv", index=False, mode='w', encoding='utf-8-sig') 
        print( c+4, "번째 파싱")

        time.sleep(1)                            
    
        
        element1 = driver.find_element_by_id("info.search.page.no{}".format(c))
        ActionChains(driver).move_to_element(element1).click(element1).perform()
        
except:
    pass

print('다했슈')
```
