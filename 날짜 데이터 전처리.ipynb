# 날짜 데이터 처리

```python
df1['날짜'] # 시리즈 (순열)
```

```python
# 문자열에서 날자 형태의 데이터로 바꿈
df1['날짜_datetime']=pd.to_datetime(df1['날짜'])
```

```python
df1.head(2)
```

```python
df1['연도']=df1['날짜_datetime'].dt.year #바꿔준 날짜형태의 데이터에서 년만 추출
```

```python
df1.head(2)
```

```python
pd.pivot_table(data=df1,index='연도',values='출고수량',aggfunc='sum')
```

# 인덱스에 추가

```python
df1['월']=df1['날짜_datetime'].dt.month
df1['주차']=df1['날짜_datetime'].dt.week
df1['일자']=df1['날짜_datetime'].dt.day
df1['요일']=df1['날짜_datetime'].dt.day_name()
```

```python
df1.head(2)
```

```python
# 피벗테이블을 이용해서 월별 출고수량 총합 출력
pd.pivot_table(data=df1 ,index='월',values='출고수량',aggfunc='sum')
```

```python
# 주차별 출고수량
pd.pivot_table(data=df1,index='주차',values='출고수량',aggfunc='sum')
```

```python
# 요일별 출고수량
# + 시각화
pd.pivot_table(data=df1,index='요일',
               values='출고수량',aggfunc='sum').plot(kind='bar')
```

```python
df2= pd.read_csv('Data02.csv')
print(df2.shape)
df2.head()
```

```python
# 문자열 타임이라 에러뜨는 구조임 
pd.to_datetime(df2['배송시작일'])
```

```python
# 위에 에러뜨는걸 해결
def date_function(x):
    return x

df2['배송시작일'].apply(date_function)
```

```python
# 인덱싱으로 추출해보기
def date_function(x):
    return x[:10]

df2['배송시작일'].apply(date_function)
```

```python
#  날짜형태로 바꿔서 집어넣기
# 배송시작일과 완료일
def date_function(x):
    return x[:10]


df2['배송시작일(datetime)']=pd.to_datetime(df2['배송시작일'].apply(date_function))
df2['배송완료일(datetime)']=pd.to_datetime(df2['배송완료일'].apply(date_function))
```

```python
df2
```

```python
# timedelta64 시간의 간격을 의미하는 데이터 타입이다
# 배송시간도 추가
df2['배송시간']=df2['배송완료일(datetime)']-df2['배송시작일(datetime)']
```

```python
df2
```

```python
# 배송시간은 describe 함수사용가능
df2['배송시간'].describe()
```

```python
# 정수형태로 인식하기 때문에 형태변환시에 마이크로 초 형태로 인지함
pd.to_datetime(df2['주문일'])
```

# 날짜에 구분기호없을시 구분기호 추가 

```python
def date_function2(x):
    result = str(x)
    return result[0:4]+'-'+result[4:6]+'-'+result[6:8]

pd.to_datetime(df2['주문일'].apply(date_function2))
```
