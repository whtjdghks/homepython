```python
import pandas as pd
# 파일 경로를 file_path 변수에 저장
file_path = 'C:/data/공공자전거 대여이력 정보_2021.06.csv'

# read csv() 함수로 데이터 프레임 변환
df1 = pd.read_csv(file_path,encoding='cp949')
df1
```

```python
# 결측값 확인
df1.isnull().sum()
```

```python
# 실제값
df1.notnull().sum()
```

```python
# 결측 데이터있는 전체 열 제거
df_drop_allcolumn = df1.dropna(axis=1)
df_drop_allcolumn
```

```python
# 특정 행 또는 결측지 제거
df_drop_allcolumn = df1.dropna(axis=0)
df_drop_allcolumn.isnull().sum()
```

```python
df2 = df1.fillna(df1.이용거리.mean())
df2.isnull().sum()
```
