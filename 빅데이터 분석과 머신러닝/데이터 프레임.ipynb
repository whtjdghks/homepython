## 데이터 연결

```python
# df.concat()
# df.append()

# 시리즈를 새로운 행으로 연결

import pandas as pd

s1 = pd.Series([0,1],index=['A','B'])
s2 = pd.Series([2,3,4],index=['c','d','e'])


s1
```

```python
# 연결 1 df.concat

test1= pd.concat([s1,s2],axis=1)  # 열에다 연결

test1.fillna(0) # 연결했을때 NaN 값 변경
```

```python
# 연결 2 df.concat

df1 = pd.DataFrame([['Dog','3'],['Bird','10'],['Tiger','6'],['Moose','3']]
                   ,index=['0','1','2','3'],columns=['종류','나이'])

# df1

df2 = pd.DataFrame([['집','0'],['초원','0'],['수풀','0'],['초원','1']],index=['0','1','2','3'],columns=['사는곳','뿔의개수'])

# df2

pd.concat([df1,df2],axis=1) # df1 과 df2를 열 기준으로 연결
```

```python
pd.concat([df1,df2],axis=0) # df1 과 df2를 행 기준으로 연결
```

## 데이터 병합

```python
# df.merge()

import pandas as pd

# 병합용 데이터 프레임

df1 = pd.DataFrame({'고객번호':[1001,1002,1003,1004,1005,1006,1007],
'이름':['강감찬','홍길동','이순신','장보고','유관순','신사임당','세종대왕']},columns=['고객번호','이름'])

df1
```

```python
df2 = pd.DataFrame({'고객번호':[1001,1001,1005,1006,1008,1001],
                    '금액':[10000,20000,15000,5000,100000,30000]},columns=['고객번호','금액'])
df2
```

```python
# inner join

pd.merge(df1,df2,on='고객번호')
pd.merge(df1,df2,how='inner',on='고객번호')
```

```python
# FUll join 방식은 키 값이 한쪽에만 있어도 데이터를 출력

test2=pd.merge(df1,df2,how='outer',on='고객번호')

test2=test2.fillna(0)
test2
```

```python
# left , right 방식은 각각 첫 번째 혹은 두 번째 데이터 프레임의 키 값을 모두 보여준다.
pd.merge(df1,df2,how='left',on='고객번호')
```

```python
test=pd.merge(df1,df2,how='right',on='고객번호')
test
```

```python
# 연습
test.sort_values('고객번호',ascending=False)
```
