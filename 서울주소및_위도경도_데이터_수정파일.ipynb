```python
# 타입에 따른 분류 시키기
def func1(row):
    if '서울 중구' in row:
        return '중구'
    elif '서울 강남구'in row:
        return '강남구'
    elif '서울 강동구'in row:
        return '강동구'
    elif '서울 강북구'in row:
        return '강북구'
    elif '서울 강서구'in row:
        return '강서구'
    elif '서울 관악구'in row:
        return '관악구'
    elif '서울 광진구'in row:
        return '광진구'
    elif '서울 구로구'in row:
        return '구로구'
    elif '서울 금천구'in row:
        return '금천구'
    elif '서울 노원구'in row:
        return '노원구'
    elif '서울 도봉구'in row:
        return '도봉구'
    elif '서울 동작구'in row:
          return '동작구'
    elif '서울 동대문구'in row:
        return '동대문구'
    elif '서울 마포구'in row:
        return '마포구'
    elif '서울 서대문구'in row:
        return '서대문구'
    elif '서울 서초구'in row:
        return '서초구'
    elif '서울 성동구'in row:
        return '성동구'
    elif '서울 성북구'in row:
        return '성북구'
    elif '서울 송파구'in row:
        return '송파구'
    elif '서울 양천구'in row:
        return '양천구'
    elif '서울 영등포구'in row:
        return '영등포구'
    elif '서울 용산구'in row:
        return '용산구'
    elif '서울 은평구'in row:
        return '은평구'
    elif '서울 종로구'in row:
        return '종로구'
    elif '서울 중랑구'in row:
        return '중랑구'
    else:
        return row
    
```

```python
# 구 라는 컬럼명으로 주소기준에 맞춰서 추가
df1['구'] = df1['주소'].apply(func1)
```
